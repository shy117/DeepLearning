### 四、频域滤波
频率域是指从函数的频率角度出发分析函数，和频率域相对的是时间域。简单说就是如果从时间域分析信号时，时间是横坐标，振幅是纵坐标。而在频率域分析的时候则是频率是横坐标，振幅是纵坐标。 举个例子，我们认为音乐是一个随着时间变化的震动。但是如果站在频域的角度上来讲，音乐是一个随着频率变化的震动，这样我们站在时间域的角度去观察你会发现音乐是静止的。同理，如果我们站在时间域的角度观察频率域的世界，就会发现世界是静止的，也是永恒的。这是因为在频率域是没有时间的概念的，那么也就没有了随着时间变化着的世界了。另外，我们需要借助傅立叶变换，才能够在得到函数在频率域中的信息

为什么要在频率域中进行图像处理？ 

(1). 可以利用频率成分和图像外表之间的对应关系。一些在空间域表述困难的增强任务，在频率域中变得非常普通；

(2). 滤波在频率域更为直观，它可以解释空间域滤波的某些性质； 

(3).可以在频率域指定滤波器，做反变换，然后在空间域使用结果滤波器作为空间域滤波器的指导

频率域滤波的步骤进行如下：

(1)给定一个大小为$M \times N$的图像 $f(x,y)$ ，然后对其进行填充，根据 $P=2M$ 和 $Q=2N$ ,填充零后得到 $f_{P}(x,y)$ 。

(2)用 $-1^{x+y}$ 乘以 $f_{p}(x+y)$,移到变换中心。

(3)计算傅里叶变换DFT得到$F(u,v)$

(4)然后进行滤波处理，得到$G(u,v)=H(u,v)F(u,v)$

(5)然后进行傅里叶反变换得到处理后的图像, $\left.g_{p}=(-1)^{x+y}\left[\text { real }\left[\mathcal{F}^{-1} G(u, v)\right]\right)\right]$

(6)获取$M \times N$的区域，得到处理后的图像 g(x,y)

#### 4.1 傅里叶变换 
看不懂，记住公式！
- 一维连续函数的傅里叶变换
    $$
    \mathrm{F}(\mathrm{u})=\frac{1}{M} \int_{-\infty}^{+\infty} \mathrm{f}(\mathrm{x}) \mathrm{e}^{-\mathrm{j} 2 \pi \mathrm{ux}} \mathrm{dx} \\ \\
    j = \sqrt{-1} 
    $$
    其中，$f(x)$表示原函数，$F(u)$表示变换之后的函数。$u$为频率域变量。

- 一维连续函数的傅里叶反变换 
    $$
    f(x)=\int_{-\infty}^{+\infty}F(u)e^{j 2 \pi ux}du
    $$

- 二维连续函数的fourier变换
    $$
    F(u, v)=\frac{1}{M N} \int_{-\infty}^{+\infty} f(x, y) e^{-j 2 \pi(u x+v y)} d x d y
    $$

- 二维连续函数的傅里叶变换
    $$
    f(x, y)=\int_{-\infty}^{+\infty} F(u, v) e^{j 2 \pi(u x+v y)} d u d v
    $$

- 二维离散傅立叶变换
    图像尺寸为$M \times N$的函数$f(x,y)$ DFT为
    $$
    f(x, y)=\int_{-\infty}^{+\infty} F(u, v) e^{j 2 \pi(u x+v y)} d u d v \\
    F(u, v)=\frac{1}{M N} \sum_{x=0}^{M-1} \sum_{y=0}^{N-1} f(x, y) e^{-j 2 \pi(u x / M+v y / N)}
    $$
    其中，$u=0,1,2,…,M-1;v=0,1,2,…,N-1$ 
    给出$F(u,v)$由反DFT反变换可得到$f(x,y)$
    $$
    f(x, y)=\sum_{u=0}^{M-1} \sum_{v=0}^{N-1} F(u, v) e^{j 2 \pi(u x / M+v y / N)}
    $$

#### 4.2 傅立叶变换的性质
- 共轭对称性
    如果$f(x,y)$是实函数，则它的傅里叶变换具有共轭对称性
    $$
    \begin{array}{l}
    F(u, v)=F^{*}(-u,-v) \\
    |F(u, v)|=|F(-u,-v)|
    \end{array}
    $$
- 周期性
    \begin{array}{l}
    F(u, v)=F(u+M, v)=F(u, v+N)=F(u+M, v+N) \\
    f(x, y)=f(x+M, y)=f(x, y+N)=f(x+M, y+N)
    \end{array}
- 平均值
    由二维傅里叶变换的定义
    $$F(u, v)=\frac{1}{M N} \sum_{x=0}^{M-1} \sum_{y=0}^{N-1} f(x, y) e^{-j 2 \pi(u x / M+v y / N)}$$

    所以 $F(0,0)=\frac{1}{M N} \sum_{x=0}^{M-1} \sum_{y=0}^{N-1} f(x, y)$
    
    而 $\bar{f}(x, y)=\frac{1}{M N} \sum_{x=0}^{M-1} \sum_{y=0}^{N-1} f(x, y) $
    
    所以在原点的傅立叶变换等于图像$f(x,y)$的平均灰度级 

#### 4.3 卷积定理
空间域和频率域的基础都是卷积定理

大小为$M \times N$的两个函数$f(x,y)$和$h(x,y)$的离散卷积
$$
f(x, y) * h(x, y)=\frac{1}{M N} \sum_{m=0}^{M-1} \sum_{n=0}^{N-1} f(m, n) h(x-m, y-n)
$$
卷积定理
$$
\begin{array}{l}
f(x, y) * h(x, y) \Leftrightarrow F(u, v) H(u, v) \\
f(x, y) h(x, y) \Leftrightarrow F(u, v) * H(u, v)
\end{array}
$$
第一个表达式表明： 
两个空间函数的卷积可以通过计算两个傅立叶变换函数的乘积的逆变换得到。 
相反，两个空间函数卷积的傅立叶变换恰好等于两个函数的傅立叶变换的乘积

- 傅里叶变换

```python
import numpy as np
import matplotlib.pyplot as plt

img = plt.imread('D:\project\Fig0424(a)(rectangle).tif')
plt.figure(figsize=(8, 8))
# 显示灰度图
plt.subplot(221), plt.imshow(img, 'gray'), plt.title('Original')

# 进行傅立叶变换，并显示结果
fft2 = np.fft.fft2(img)
plt.subplot(222), plt.imshow(np.abs(fft2), 'gray'), plt.title('FFT')

# 将图像变换的原点移动到频域矩形的中心，并显示效果
shift2center = np.fft.fftshift(fft2)
plt.subplot(223), plt.imshow(np.abs(shift2center), 'gray'), plt.title('FCenter')

# 对中心化后的结果进行对数变换，并显示结果
log_shift2center = np.log(1 + np.abs(shift2center))
plt.subplot(224), plt.imshow(log_shift2center, 'gray'), plt.title('LogCF')
plt.show()
```

- 低通滤波器


```python
im = plt.imread('D:\project\Fig0429(a)(blown_ic).tif')
H = im.shape[0]
W = im.shape[1]
P = H * 2
Q = W * 2
Pshape = [P, Q]
# 原始图像
plt.figure(figsize=(8, 8))
P1 = im
plt.subplot(341)
plt.imshow(P1, 'gray')
plt.title('Original')

# 填充后的图像
P2 = np.zeros(Pshape, dtype=int)
for i in range(H):
    for j in range(W):
        P2[i, j] = P1[i, j]
plt.subplot(342)
plt.imshow(P2, 'gray')
plt.title('Padding')

plt.subplot(343)
plt.imshow(np.abs(P2), 'gray')
plt.title('Original')

plt.subplot(344)
P4 = np.fft.fft2(P2)
P4 = np.fft.fftshift(P4)
plt.imshow(np.log(1 + np.abs(P4)), 'gray')
plt.title('FFT')


def ButterworthLowpassFilter(shape, d0, n):
    H = np.zeros(shape, dtype=float)
    IH = shape[0]
    IW = shape[1]
    for i in range(IH):
        for j in range(IW):
            d = np.sqrt((i - IH / 2) ** 2 + (j - IW / 2) ** 2)
            H[i, j] = 1 / (1 + (d / d0) ** (2 * n))
    return H


def ButterworthHighpassFilter(shape, d0, n):
    H = ButterworthLowpassFilter(shape, d0, n)
    return 1 - H


def IdealLowpassFilter(shape, d0):
    H = np.zeros(shape, dtype=float)
    IH = shape[0]
    IW = shape[1]
    for i in range(IH):
        for j in range(IW):
            d = np.sqrt((i - IH / 2) ** 2 + (j - IW / 2) ** 2)
            if d <= d0:
                H[i, j] = 1
            else:
                H[i, j] = 0
    return H


def IdealHighpassFilter(shape, d0):
    H = IdealLowpassFilter(shape, d0)
    return 1 - H


def GaussLowpassFilter(shape, d0):
    H = np.zeros(shape, dtype=float)
    IH = shape[0]
    IW = shape[1]
    for i in range(IH):
        for j in range(IW):
            d = np.sqrt((i - IH / 2) ** 2 + (j - IW / 2) ** 2)
            H[i, j] = np.exp(-d * d / (2 * d0 * d0))
    return H


def GaussHighpassFilter(shape, d0):
    H = GaussLowpassFilter(shape, d0)
    return 1 - H

H = ButterworthLowpassFilter(P4.shape, 50, 1)
plt.subplot(345)
plt.imshow(H, 'gray')
plt.title('LowpassFilter')

P5 = P4 * H
plt.subplot(346)
plt.imshow(np.abs(P5), 'gray')
plt.title('logHF Spectrum')

plt.subplot(347)
newImage = np.fft.ifftshift(P5)
newImage = np.fft.ifft2(newImage)
plt.imshow(np.abs(newImage), 'gray')
plt.title('IFFT')

plt.subplot(348)
result = newImage[0:im.shape[0], 0:im.shape[1]]
plt.imshow(np.abs(result), 'gray')
plt.title('Result')

plt.subplot(349)
H_ = 1 - H
plt.imshow(np.abs(H_), 'gray')
plt.title('Highpass Filter')

plt.subplot(3,4, 10)
H_ = 1 - H
P6 = H_ * P4
plt.imshow(np.abs(P6),'gray')
plt.title('HF Spectrum')


plt.subplot(3, 4, 11)
newImage = np.fft.ifftshift(P6)
newImage = np.fft.ifft2(newImage)
plt.imshow(np.abs(newImage), 'gray')
plt.title('IFFT')

plt.subplot(3, 4, 12)
result = newImage[0:im.shape[0], 0:im.shape[1]]
plt.imshow(np.abs(result), 'gray')
plt.title('Result')

plt.show()
```

- 带通滤波器

```python
m = plt.imread('D:\project\Fig0429(a)(blown_ic).tif')
plt.figure(figsize=(16, 8))

def IdealBRFilter(shape, D0, W):
    H = np.zeros(shape, dtype=float)
    for i in range(shape[0]):
        for j in range(shape[1]):
            d = np.sqrt((i - shape[0] / 2) ** 2 + (j - shape[1] / 2) ** 2)
            if D0 - W / 2 <= d <= D0 + W / 2:
                H[i, j] = 0
            else:
                H[i, j] = 1
    return H


def IdealBPFilter(shape, D0, W):
    H = IdealBRFilter(shape, D0, W)
    return 1 - H


def ButterworthBRFilter(shape, D0, W, n):
    H = np.zeros(shape, dtype=float)
    for i in range(shape[0]):
        for j in range(shape[1]):
            d = np.sqrt((i - shape[0] / 2) ** 2 + (j - shape[1] / 2) ** 2)
            H[i, j] = 1 / (1 + (d * W / (d * d - D0 * D0 + 0.00001)) ** (2 * n))
    return H


def ButterworthBPFilter(shape, D0, W, n):
    H = ButterworthBRFilter(shape, D0, W, n)
    return 1 - H


def GaussBRFilter(shape, D0, W):
    H = np.zeros(shape, dtype=float)
    for i in range(shape[0]):
        for j in range(shape[1]):
            d = np.sqrt((i - shape[0] / 2) ** 2 + (j - shape[1] / 2) ** 2)
            H[i, j] = 1 - np.exp(-((d * d - D0 * D0) / (d * W + 0.00001)) ** 2)
    return H


def GaussBPFilter(shape, D0, W):
    H = GaussBRFilter(shape, D0, W)
    return 1 - H


def ImagePad(im):
    id = np.pad(im, ((0, im.shape[0]), (0, im.shape[1])))
    return id


def Filter(im, H):
    IH = im.shape[0]
    IW = im.shape[1]
    P1 = im
    plt.subplot(2, 4, 1)
    plt.imshow(P1[0:int(IH / 2), 0:int(IW / 2)], 'gray')
    plt.title('Original')

    plt.subplot(2, 4, 2)
    plt.imshow(P1, 'gray')
    plt.title('Padding')

    plt.subplot(2, 4, 3)
    plt.imshow(H, 'gray')
    plt.title('Filter')

    plt.subplot(2, 4, 4)
    P2 = np.fft.fft2(P1)
    P2_ = np.fft.fftshift(P2)
    plt.imshow(np.log(1 + np.abs(P2_)), 'gray')
    # plt.imshow(np.abs(P2_), 'gray')
    plt.title('logFFT spectrum')

    plt.subplot(2, 4, 5)
    P3 = P2_ * H
    P3_ = np.fft.ifftshift(P3)
    P3_ = np.fft.ifft2(P3_)
    plt.imshow(np.abs(P3_), 'gray')
    plt.title('IFFT')

    plt.subplot(2, 4, 6)
    P4 = np.abs(P3_)[0:int(IH / 2), 0:int(IW / 2)]
    plt.imshow(P4, 'gray')
    plt.title('Result')

    plt.subplot(2, 4, 7)
    P5 = P2_ * (1 - H)
    P5_ = np.fft.ifftshift(P5)
    P5_ = np.fft.ifft2(P5_)
    plt.imshow(np.abs(P5_), 'gray')
    plt.title('IFFT')

    plt.subplot(2, 4, 8)
    P6 = np.abs(P5_)[0:int(IH / 2), 0:int(IW / 2)]
    plt.imshow(P6, 'gray')
    plt.title('Result')
    plt.show()


ImageP = ImagePad(im)
H = GaussBRFilter(ImageP.shape, 100, 20)
Filter(ImageP, H)
```
